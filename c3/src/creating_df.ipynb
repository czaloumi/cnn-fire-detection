```
fires18 = drive.CreateFile({'id': '1MUFRzAX-RmO_tiyjae2cw7yZ4w-ioauL'})
fires18.GetContentFile('2018_fires.csv')
```

```
fires18 = pd.read_csv('2018_fires.csv')
notes18 = fires18.pop('Notes')
fires18['Start date'] = pd.to_datetime(fires18['Start date'], format='%B %d, %Y')
fires18['Start date'] = fires18['Start date'].apply(lambda x: x.strftime('%-m/%-d/%Y'))
fires18['Containment date'] = pd.to_datetime(fires18['Containment date'], format='%B %d, %Y')
fires18['Containment date'] = fires18['Containment date'].apply(lambda x: x.strftime('%-m/%-d/%Y'))
```

```
fires18[['Start date','Containment date']] = fires18[['Start date','Containment date']].apply(pd.to_datetime)
fires2018 = fires18.loc[fires18.index.repeat((fires18['Containment date'] - fires18['Start date']).dt.days + 1)]
fires2018['Date'] = fires2018['Start date'] + pd.to_timedelta(fires2018.groupby(level=0).cumcount(), unit='d')
fires2018 = fires2018.reset_index(drop=True)
fires2018.drop(['Start date', 'Containment date'], axis=1, inplace=True)
fires2018['Date'] = pd.to_datetime(fires2018['Date'], format='%B %d, %Y')
fires2018['Date'] = fires2018['Date'].apply(lambda x: x.strftime('%-m/%-d/%Y'))
```
